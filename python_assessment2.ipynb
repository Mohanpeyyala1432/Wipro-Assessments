1. Create a program that asks the user for 5 numbers and prints the maximum and minimum using loops.

```python
#count variable to count the numbers while adding into the list
count=0

# empty list to store the numbers
numbers=[]

# using while for taking the numbers from the user
while count<5:
  num=int(input("enter the number "))

   # adding numbers into the list
  numbers.append(num)
  count+=1

maximum=numbers[0]
minimum=numbers[0]

#iterating the loop
for i in numbers:

  #for maximum element
  if i>maximum:
    maximum=i

  #for minimum element
  if i<minimum:
    minimum=i

#printing the elements
print("maximum  number is ", maximum)
print("minimum number is ", minimum)
```

2.Write a program to check if a given number is prime using a loop.

4.Given a list of numbers, count how many are even and how many are odd.

```python
#list of numbers
numbers=[3,5,7,4,8,9,1]

#for even numbers
even_count=0

#for odd numbers
odd_count=0

for i in numbers:

  #if condition for even numbers
  if i%2==0:
    even_count+=1 #incrementing the count

  #else condition for odd numbers
  else:
    odd_count+=1 #incrementing the count

print("total even numbers in a list ",even_count)
print("total odd numbers in a list ", odd_count)
```

17.Write a program to count frequency of each element in a list.

```python
elements = [4,3,5,4,3,6,7,2,3,2,8,7,9]
visited = []

for i in elements:
    if i not in visited:
        count = 0
        for j in elements:
            if i == j:
                count += 1

        print("frequency of", i, "is", count)
        visited.append(i)
```
